#### 2.Vanishing Gradient Problem
The vanishing gradient problem happens when gradients (partial derivatives of the loss w.r.t. weights) become very small as they’re backpropagated through many layers. As a result, earlier (lower) layers learn extremely slowly or stop learning altogether.

z=W.x+b equation pass through to the activation function.

Activation function:

Activation functions are used in deep learning to introduce non-linearity, enabling neural networks to learn and model complex patterns beyond simple linear relationships.

In other word , activation function says that weather the neuron should activate or not.


#### Types of activation function


🔹 1. Linear Activation

*   Rarely used except in the output layer for regression tasks.

🔹 2. Sigmoid (Logistic)
    

*   Output range: (0, 1)

*   Used for probabilities or binary classification.
*   Prone to vanishing gradients.


🔹 3. Tanh (Hyperbolic Tangent)


*   Output range: (-1, 1)

*   Zero-centered but still can suffer vanishing gradients.


🔹 4. ReLU (Rectified Linear Unit)
Formula:

* f(x)=max(0,x)

* Most common in hidden layers.

* Fast and reduces vanishing gradient problem but can cause “dead” neurons.

🔹 5. Leaky ReLU

* Formula: f(x)=x if 𝑥>0, else
αx

* Fixes ReLU’s dead neuron problem by allowing a small slope for negative values.

🔹 6. ELU (Exponential Linear Unit)

* Like Leaky ReLU but smoother for negative values.

* Helps faster convergence.

🔹 7. Softmax

* Converts raw scores into probabilities for multi-class classification.

* Usually used in the output layer.

🔹 8. Swish / GELU

* Newer activations combining properties of sigmoid and ReLU.

* Often improve performance in deep networks like transformers.

🔹 9. pre ReLu


## Which Activation function we should use

## ReLu:
default choice for(both Regression and Classification) hidden layers in most deep nets.

Why: avoids vanishing gradients (for
𝑥>0), simple and fast.

## Sigmoid:
Use when: output layer of a binary classification model (probability between 0 and 1)

Why not hidden layers: saturates at 0 or 1, causing vanishing gradients.

## SoftMax:
Use when: output layer for multi-class classification (probabilities across classes sum to 1).

Not for hidden layers.

## Leaky ReLU / ELU / GELU:

Use when: you like ReLU but want to reduce the “dead neuron” issue.

## Linear Activation function:
use when: use in output Layers for Regression

## Optimizer( use to reduce the error and helps to converge )


🔹Type 1: Basic Gradient Descent Variants

| Optimizer                             | Idea                                            | When to Use                       |
| ------------------------------------- | ----------------------------------------------- | --------------------------------- |
| **1.Batch Gradient Descent**            | Uses all training data for each update.         | Small datasets; stable but slow.  |
| **2.Stochastic Gradient Descent (SGD)** | Updates weights per sample.                     | Large datasets; faster but noisy. |
| **3.Mini-Batch SGD**                    | Updates weights on small batches (most common). | Default for deep learning.        |

1.Batch Gradient Descent

 Advantage:


* Convergence will happen fast

Disadvantage:


*   As it use all training data for each updation, that is huge resource intensive task because of we need high cpu,gpu etc.


2.Stochastic Gradient Descent

 It takes one data point for each iteratio or updation
 example: consider, we have a dataset of 50000 datapoint. so for each epoch 50000 time updation/iteration will happen one by one.

  Advantage:

 * Unlike Gradient Descent it solve the resource issue.
Disadvantage:


*   As it use 1 data point(1 training data) for each epoch , thats leads to late convergence.

*   Time Complexity is high.
* Noise get introduced.


3.Mini-Batch SGD

we use a small batch of data points (like32, 64, 128) for each update.
This is why it’s called mini-batch SGD.

Advantage:
* Convergence speed will increase
* Noise will be less compare to SGD
* Efficient in terms of resource usage

Disadvantage:
*  Noise still exists


🔹Type 2: Momentum-based Methods

| Optimizer                               | Idea                                                | Benefit                                  |
| --------------------------------------- | --------------------------------------------------- | ---------------------------------------- |
| **1.SGD With Momentum**                      | Adds a “velocity” term to smooth updates.           | Faster convergence, avoids local minima. |
| **2.Nesterov Accelerated Gradient (NAG)** | Like momentum but looks ahead at the next position. | More responsive than vanilla momentum.    |


1.SGD With Momentum

It introduced Exponential weighted average

Advantage:
* Reduce the Noise by smoothning the line on gradient descent
* Quick Convergence

🔹 Type 3: Adaptive Learning Rate Methods

| Optimizer   | Idea                                                                 | Benefit                                                |
| ----------- | -------------------------------------------------------------------- | ------------------------------------------------------ |
| **1.Adagrad** | Adapts learning rate per parameter (larger for infrequent features). | Good for sparse data; learning rate shrinks over time. |
| **2.Adadelta and RMSProp** | Uses moving average of squared gradients to normalize.               | Handles non-stationary objectives better than Adagrad. |


1.Adagrad/ adaptive GD

Standard gradient descent uses one fixed learning rate
𝜂
η for all parameters.

But in many problems (especially sparse data like text or recommender systems), some parameters get updated very often, others rarely.
It’s inefficient to use the same learning rate for all.
Adaptive optimizers adjust the learning rate per parameter based on past gradients.

In other word, learning rate changes as the convergence happen

Disadvantage:
* Learning rate decays continuously; can become extremely small and stop learning (Possibility to become zero).That means no updation will happen.


2.Adadelta and RMSProp

Advantage:
* Prevent possibility of learning rate to become zero ( Dynamic but not becomes Zero)
* Smoothning the line using exponential weighted average


🔹 Type 4: Adaptive + Momentum
| Optimizer                             | Idea                                                              | Benefit                                        |
| ------------------------------------- | ----------------------------------------------------------------- | ---------------------------------------------- |
| **1.Adam (Adaptive Moment Estimation)** | Combines RMSProp + Momentum. Tracks mean & variance of gradients. | Default choice for many deep learning tasks.   

1. Adam optimizer

SGD with Momentum+ RMSProp( dynamic Learning rate+ smoothning factor )

### Comparison table of all optimizer

| **Optimizer**                           | **Type / Idea**                                                                                      | **Advantages**                                                                                     | **Disadvantages**                                                                                     |
| --------------------------------------- | ---------------------------------------------------------------------------------------------------- | -------------------------------------------------------------------------------------------------- | ----------------------------------------------------------------------------------------------------- |
| **SGD (Stochastic Gradient Descent)**   | Updates weights using a small batch (or one sample) at a time                                        | – Simple and widely used<br>– Good generalization<br>– Low memory usage                            | – Slow convergence<br>– Can get stuck in local minima / saddle points<br>– Sensitive to learning rate |
| **SGD + Momentum**                      | Adds an exponential moving average of past gradients to the update (momentum term)                   | – Faster convergence than plain SGD<br>– Helps escape local minima<br>– Smooths updates            | – Requires tuning of momentum parameter<br>– Still sensitive to learning rate                                                        |
| **Adagrad**                             | Per-parameter adaptive learning rate, scaled by the sum of squared gradients                         | – No manual learning rate scheduling<br>– Handles sparse data well                                 | – Learning rate decays too much over time<br>– May stop learning prematurely                          |
| **RMSProp**                             | Modifies Adagrad with an exponential moving average of squared gradients (to avoid aggressive decay) | – Handles non-stationary objectives<br>– Works well for RNNs<br>– Keeps learning rate more stable  | – Hyperparameters (decay rate) need tuning<br>– Still no bias correction                              |
| **Adadelta**                            | Extension of RMSProp that also uses running average of parameter updates (no manual LR needed)       | – Reduces need for learning rate tuning<br>– More robust to large gradients                        | – Can still be slower than Adam<br>– Convergence not always as strong                                 |
| **Adam** (Adaptive Moment Estimation)   | Combines RMSProp + momentum (1st & 2nd moments of gradients)                                         | – Fast convergence<br>– Works well in practice for most problems<br>– Less sensitive to initial LR | – Can sometimes generalize worse than SGD<br>– May converge to sharp minima
