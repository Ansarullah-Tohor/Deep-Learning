#### Forward propagation in RNN
Recurrent Neural Networks (RNNs) are designed to process sequential data by maintaining a memory of previous inputs through their hidden states. Forward propagation in RNNs is the process of passing input data through the network to compute outputs at each time step, leveraging both the current input and the hidden state from the previous step.


Applications of RNN Forward Propagation

Language Modeling: Predicting the next word or character in a sequence.

Sentiment Analysis: Classifying text sequences as positive or negative.

Time-Series Forecasting: Predicting future values based on past data.

Speech Recognition: Processing audio sequences to generate text.

**LSTM (Long Short-Term Memory): A special kind of RNN designed to solve the vanishing/exploding gradient problem, so it can learn long-term dependencies better than basic RNNs.**


**When sequences are long, the gradients during training can vanish or explode. **

**To prevent this vanishing gradient problem we have to use LSTM (Long Short Term Memory) RNN**


LSTM RNN = an RNN whose units are LSTM cells.


### Why LSTM was invented

Normal RNNs quickly “forget” information as the sequence gets long.
LSTM adds an internal “memory cell” and gating mechanisms so it can:

Remember important info for a long time

Forget irrelevant info automatically

Control the flow of information in and out of the cell


### When to use LSTM

* Time-series prediction(stock price forecasting,weather prediction)

* Natural language processing (text classification, machine translation)

* Speech recognition

* Any problem where past information matters for current prediction

* video Analysis (Understanding sequential frames in video data)

#### Key Features of LSTMs
  1.Memory Cells:
   LSTMs have memory cells that store information over time, enabling them to learn long-term dependencies.

  2.Gates: LSTMs use three types of gates to control the flow of information:
  * Forget Gate: Decides what information to discard from the cell state.
  * Input Gate: Determines what new information to store in the cell state.
  * Output Gate: Controls what information to output from the cell state.

  3.Gradient Flow: By maintaining a constant error flow through the memory cells, LSTMs mitigate the vanishing gradient problem.
